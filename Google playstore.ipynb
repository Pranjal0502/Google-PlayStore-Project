```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('P:/Unified mentor/googleplaystore.xlsx')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
df.dropna(subset = ['App','Type','Content Rating','Current Ver','Android Ver'], inplace = True)
```

```python
df
```

```python
#Remove commas and the '+' sign, then convert to integer
df['Installs'] = df['Installs'].astype(str).str.replace(',', '').str.replace('+', '').astype(int)

#Group by Category, sum the installs and sort values by installs
installs_by_category = df.groupby('Category')['Installs'].sum().reset_index().sort_values(by='Installs', ascending=False).head(20)

#Plot the data using Seaborn
plt.figure(figsize=(10, 8))
sns.barplot(x='Installs', y='Category', data=installs_by_category)

# Add title and labels
plt.title('Total Installs by Category', fontsize=16)
plt.xlabel('Total Installs', fontsize=14)
plt.ylabel('Category', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

From the above graph we can see that mostly Game category has been installed.

```python
app_reviews = df.groupby('App')['Reviews'].sum().reset_index().sort_values(by='Reviews', ascending=False).head(5)

plt.figure(figsize=(10, 8))
sns.barplot(x='Reviews', y='App', data=app_reviews, palette='coolwarm')

plt.title('Top 5 Most Reviewed Apps', fontsize=16)
plt.xlabel('Total Reviews', fontsize=14)
plt.ylabel('App', fontsize=14)
plt.show()
```

The above graph shows that the most reviewed App is Instagram.

```python
android_ver_counts = df['Android Ver'].value_counts().reset_index().head(10)
android_ver_counts.columns = ['Android Ver', 'Count']

plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Android Ver', data=android_ver_counts)
plt.title('Most Commonly Supported Android Versions', fontsize=16)
plt.xlabel('Number of Apps', fontsize=12)
plt.ylabel('Android Version', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
ratings_by_genres = df.groupby('Genres')['Rating'].sum().reset_index().sort_values(by='Rating', ascending=False).head(10)

#Plot the data using Seaborn
plt.figure(figsize=(10, 8))
sns.barplot(x='Rating', y='Genres', data=ratings_by_genres)

# Add title and labels
plt.title('Total Rating by Genres', fontsize=16)
plt.xlabel('Total Rating', fontsize=14)
plt.ylabel('Genres', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of Free and Paid apps
app_types = df['Type'].value_counts()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(app_types, labels=app_types.index, autopct='%1.1f%%')
plt.legend(labels=app_types.index, title="App Type")

# Add a title
plt.title('Distribution of Free vs Paid Apps',fontsize = 14)

# Show the plot
plt.show()
```

```python
# Count the occurrences of Category
category = df['Category'].value_counts().head(10)

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(category, labels=category.index, autopct='%1.1f%%')

# Add a title
plt.title('Distribution of Category',fontsize = 14)

# Show the plot
plt.show()
```

```python
df_clean = df.dropna(subset=['Size', 'Installs']).head(20)

# Scatter plot to visualize Size vs. Installs
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Size', y='Installs', data=df_clean)

# Add title and labels
plt.title('App Size vs. Installs',fontsize = 16)
plt.xlabel('App Size (Bytes)',fontsize = 14)
plt.ylabel('Number of Installs',fontsize = 14)

# Show the plot
plt.show()
```
